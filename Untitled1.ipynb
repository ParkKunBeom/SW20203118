```python
from turtle import *
import random

# 키보드 이벤트
# 왼쪽 이동
def key_left():
    my.seth(180)
    my.fd(step)
    if my.xcor() < -300:
        my.setx(-300)

# 오른쪽 이동
def key_right():
    my.seth(0)
    my.fd(step)
    if my.xcor() > 300:
        my.setx(300)

# 위쪽 이동
def key_up():
    my.seth(90)
    my.fd(step)
    if my.ycor() > 300:
        my.sety(300)

# 아래쪽 이동
def key_down():
    my.seth(270)
    my.fd(step)
    if my.ycor() < -300:
        my.sety(-300)
```

```python
# 모양 랜덤 선택하는 함수
def make_shape(size):
    shape = random.choice(["circle", "square", "triangle"])
    return shape

# 최초 장애물 생성
def make_obstacle(size):
    x = random.randint(-200, max_x)
    y = random.randint(min_y, 200)
    color = random.choice(["red", "green", "blue", "yellow", "pink", "purple"])
    while True:
        dx = random.randint(-1, 1)  # 이동벡터
        dy = random.randint(-1, 1)  # 이동벡터
        if dx != 0 or dy != 0:
            break
    shape = make_shape(size)
    return [x, y, size, shape, color, dx * ball_speed, dy * ball_speed]  # 속도 적용
```

```python
# 거리 계산 함수, 피터고라스의 정리 이용
def distance(x1, y1, x2, y2):
    return ((x2 - x1) ** 2 + (y2 - y1) ** 2) ** 0.5

# 움직이는 장애물
def animate_obstacle(pen, obstacle_list):
    tracer(False)
    game = True
    death = 0
    while game:
        listen()
        pen.clear()  # 장애물을 지움
        # 장애물의 이동
        for i in range(len(obstacle_list)):
            obs_x, obs_y, size, shape, color, dx, dy = obstacle_list[i]
            obs_x += dx
            obs_y += dy
            obstacle_list[i][0] = obs_x
            obstacle_list[i][1] = obs_y
            # 거북이와 장애물의 충돌 -> 데스 카운트 +1
            if distance(my.xcor(), my.ycor(), obs_x, obs_y) <= 15:
                my.setpos(-goal, goal)
                death += 1
            # 장애물의 좌표가 화면을 벗어날 시 이동 벡터에 -1 값을 곱하여 반대방향으로 움직이도록 함
            if obs_x < min_x or obs_x > max_x:
                obstacle_list[i][5] *= -1
            if obs_y < min_y or obs_y > max_y:
                obstacle_list[i][6] *= -1
            # 장애물과 장애물의 충돌 판정
            # 중심 좌표 간 거리를 이용하여 충돌 판정을 구현
            for j in range(len(obstacle_list)):
                if not i == j:
                    obs2_x = obstacle_list[j][0]
                    obs2_y = obstacle_list[j][1]
                    if distance(obs_x, obs_y, obs2_x, obs2_y) <= ball_size:
                        obstacle_list[i][5] *= -1
                        obstacle_list[i][6] *= -1
                        obstacle_list[j][5] *= -1
                        obstacle_list[j][6] *= -1
                        obstacle_list[i][0] += obstacle_list[i][5] * 2
                        obstacle_list[i][1] += obstacle_list[i][6] * 2
                        obstacle_list[j][0] += obstacle_list[j][5] * 2
                        obstacle_list[j][1] += obstacle_list[j][6] * 2
            pen.setpos(obs_x, obs_y)
            pen.shape(shape)
            pen.color(color)
            pen.stamp()
        # 목표지점 도착
        if goal <= my.xcor() <= goal + 20 and -goal - 20 <= my.ycor() <= -goal:
            game = False
            return death
        update()  # 스크린 갱신
```

```python
#메인문

ballnumber = input("장애물의 개수: ")
Ballnumber = int(ballnumber)
global min_x, max_x, min_y, max_y, ball_size, goal, step, ball_speed
# 기본 설정 (창의 크기, 볼의 크기, 목표 위치, 이동 거리)
max_x = 300
min_x = -max_x
max_y = 300
min_y = -max_y
ball_size = 20
goal = 250
step = 20

ball_speed = float(input("원하는 장애물의 속도를 입력하세요 (0.1 ~ 2.0): "))  # 사용자 입력으로 장애물의 속도 조절
if ball_speed < 0.1:
    ball_speed = 0.1
elif ball_speed > 2.0:
    ball_speed = 2.0

setup(max_x * 2, max_y * 2)


global my
my = Turtle()
my.up()
# 도착지점을 그림
my.setpos(goal, -goal)
my.down()
my.seth(0)
my.fd(20)
my.seth(270)
my.fd(20)
my.seth(180)
my.fd(20)
my.seth(90)
my.fd(20)
my.up()

# 기본 설정
my.shape("turtle")
my.setpos(-goal, goal)
# my.down()
onkey(key_right, "Right")
onkey(key_left, "Left")
onkey(key_up, "Up")
onkey(key_down, "Down")

my_pen = Turtle()
my_pen.up()
my_pen.ht()
obstacle_list = []
# 장애물 생성 정보를 받아와 리스트에 추가
for i in range(Ballnumber):
    obstacle = make_obstacle(ball_size)
    obstacle_list.append(obstacle)
deathCount = animate_obstacle(my_pen, obstacle_list)  # 에니메이션
print("Goal in! Death Count : %d" % (deathCount))
```
